```python
with open("earnings.txt", "r") as f:
    data = [s.strip() for s in f.readlines()]

data_tuples = [el.split(":") for el in data]
data_tuples = [(spl[0].strip(), spl[1].strip()) for spl in data_tuples]
data_tuples.sort(key=lambda x: x[1])
print(data_tuples)
```
